```python
#import libraries
import statsmodels.api as sm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Load Data
df = pd.read_csv("diamond.csv")
df.head()
```

# Data preparation

```python
#isolate X and Y
y = df.price
X = df.carat
```

```python
#Adding constant to X
X = sm.add_constant(X)
X.head()
```

#Linear Regression

```python
#linear regression
model1 = sm.OLS(y, X).fit()
print(model1.summary())
```

```python
#plotting
X = df.carat
plt.plot(X, y, 'o')
b, a = np.polyfit(X, y, 1)
plt.plot(X, b*X + a)
plt.xlabel("Carat")
plt.ylabel("Price")
```
