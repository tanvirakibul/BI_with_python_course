```python
#libraries 
import pandas as pd
import statsmodels.api as sm
import numpy as np
import seaborn as sns
```

```python
#Loading the data
df = pd.read_csv("Clothing.csv")
df.head()
```

```python
#picking variables
df = df[["tsales", "margin", "nown", "inv1", "inv2", "ssize", "start"]]
df.head(1)
```

#data analysis

```python
#Summary statistics
df.describe()
```

```python
#Correlation Matrix
sns.heatmap(df.corr(),
            annot = True,
            fmt = '.1g',
            center = 0,
            cmap = "coolwarm",
            linewidths = 1,
            linecolor = 'black')
```

#Multilinear Regression prep

```python
#Prepare X and Y
y = df.tsales
X = df.drop(columns = ["tsales"])
X = sm.add_constant(X)
```

```python
#Training and Test Set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.2,
                                                    random_state = 1502)
```

# Multilinear Regression

```python
#Regression
model = sm.OLS(y_train, X_train).fit()
print(model.summary())
```

```python
#predictions
predictions = model.predict(X_test)
predictions[:2]
```

```python
#accuracy assessment
from sklearn.metrics import mean_squared_error, mean_absolute_error
print("MAE:", round(mean_absolute_error(y_test, predictions), 0))
print("RMSE:", round(np.sqrt(mean_squared_error(y_test, predictions)), 0))
```
